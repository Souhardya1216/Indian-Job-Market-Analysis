```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/indian-fresher-job-market/job_market.csv')
```

```python
df['Monthly Salary'].fillna(df['Monthly Salary'].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
df['Locality'].fillna("Unknown", inplace=True)
```

```python
df.isna().sum()
```

```python
 #Replace ₹ and commas, then extract numeric values
df['Salary (Cleaned)'] = df['Salary'].str.replace("₹", "", regex=True).str.replace(",", "", regex=True)
```

```python
# Split salary ranges and take the mean
df['Salary (Cleaned)'] = df['Salary (Cleaned)'].str.split("-").apply(lambda x: np.mean([float(i) for i in x if i.strip().replace(".", "").isdigit()]) if isinstance(x, list) and len(x) > 0 else np.nan)
```

```python
# Fill missing values with median salary
df['Salary (Cleaned)'].fillna(df['Salary (Cleaned)'].median(), inplace=True)
```

```python
df[['Salary', 'Salary (Cleaned)']].head()
```

```python
# Exploratory Data Analysis (EDA)
plt.figure(figsize=(12, 6))
sns.histplot(df['Salary (Cleaned)'], bins=30, kde=True)
plt.title("Distribution of Salaries")
plt.xlabel("Salary (in INR)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Most Common Job Titles
plt.figure(figsize=(12, 6))
df['Job Title'].value_counts().nlargest(10).plot(kind='bar')
plt.title("Top 10 Most Common Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Salaries by State
plt.figure(figsize=(14, 6))
df.groupby("State")["Salary (Cleaned)"].median().sort_values(ascending=False).plot(kind='bar')
plt.title("Median Salary by State")
plt.xlabel("State")
plt.ylabel("Median Salary (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Job Listings by City
plt.figure(figsize=(14, 6))
df['City'].value_counts().nlargest(10).plot(kind='bar')
plt.title("Top 10 Cities with Most Job Listings")
plt.xlabel("City")
plt.ylabel("Number of Job Listings")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Set a general style for the plots
sns.set_style("whitegrid")

# Distribution of Salaries with Mean and Median Lines
plt.figure(figsize=(12, 6))
sns.histplot(df['Salary (Cleaned)'], bins=30, kde=True, color='blue', alpha=0.6)
plt.axvline(df['Salary (Cleaned)'].mean(), color='red', linestyle='dashed', linewidth=2, label=f"Mean: {df['Salary (Cleaned)'].mean():,.0f}")
plt.axvline(df['Salary (Cleaned)'].median(), color='green', linestyle='dashed', linewidth=2, label=f"Median: {df['Salary (Cleaned)'].median():,.0f}")
plt.title("Distribution of Salaries")
plt.xlabel("Salary (INR)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
# Top 10 Most Common Job Titles
plt.figure(figsize=(14, 6))
df['Job Title'].value_counts().nlargest(10).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Top 10 Most Common Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Number of Listings")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Median Salary by State
plt.figure(figsize=(14, 6))
df.groupby("State")["Salary (Cleaned)"].median().sort_values(ascending=False).plot(kind='bar', color='coral', edgecolor='black')
plt.title("Median Salary by State")
plt.xlabel("State")
plt.ylabel("Median Salary (INR)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Cities with Most Job Listings (Horizontal Bar Chart for Readability)
plt.figure(figsize=(14, 6))
df['City'].value_counts().nlargest(10).plot(kind='barh', color='lightgreen', edgecolor='black')
plt.title("Top 10 Cities with Most Job Listings")
plt.xlabel("Number of Job Listings")
plt.ylabel("City")
plt.show()
```

```python
# Salary Comparison Across Different Job Titles (Boxplot)
plt.figure(figsize=(14, 6))
top_job_titles = df['Job Title'].value_counts().nlargest(10).index  # Top 10 most common job titles
sns.boxplot(data=df[df['Job Title'].isin(top_job_titles)], x='Job Title', y='Salary (Cleaned)', palette="coolwarm")
plt.xticks(rotation=45, ha="right")
plt.title("Salary Distribution Across Top 10 Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Salary (INR)")
plt.show()
```

```python
# Job Listings by State (Heatmap-style Bar Chart)
plt.figure(figsize=(14, 6))
state_counts = df['State'].value_counts().sort_values(ascending=False)
sns.barplot(x=state_counts.index, y=state_counts.values, palette="viridis", edgecolor="black")
plt.xticks(rotation=45, ha="right")
plt.title("Number of Job Listings by State")
plt.xlabel("State")
plt.ylabel("Number of Job Listings")
plt.show()
```

```python
# Correlation Between Monthly Salary and Cleaned Salary (Scatter Plot)
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['Monthly Salary'], y=df['Salary (Cleaned)'], alpha=0.6, color='purple')
plt.title("Correlation Between Monthly Salary and Cleaned Salary")
plt.xlabel("Monthly Salary (INR)")
plt.ylabel("Cleaned Salary (INR)")
plt.show()
```
